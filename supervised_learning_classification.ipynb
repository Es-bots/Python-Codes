# Fit

```python
from sklearn.neighbors import KNeighborsClassifier 

knn = KNeighborsClassifier(n_neighbors = 6
                           , metric = 'minkowski', p=2) 

x = iris.data
y = iris.target

knn.fit(x, y) 
```

# Evaluation model

```python
knn = KNeighborsClassifier(n_neighbors=5) 
knn.fit(x_train, y_train) 
y_predict = knn.predict(x_test)
print(y_predict) 


knn.score(x_test, y_test) 
```

# Over Fitting and Under Fitting

```python
neighbors = np.arange(1, 30) 

train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for k in neighbors:
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(x_train, y_train)
    train_accuracy[k-1] = knn_model.score(x_train, y_train)
    test_accuracy[k-1] = knn_model.score(x_test, y_test)
    
plt.plot(neighbors, train_accuracy, label = 'train accuracy')
plt.plot(neighbors, test_accuracy, label = 'test accuracy')
plt.legend()
plt.xlabel('Number of neigbors')
plt.ylabel('accuracy')
plt.annotate(xy = (2, 0.91), s = 'overfitting', xytext=(5, 0.92), 
             arrowprops={'facecolor' : 'silver', 'width' : 4})
plt.annotate(xy = (28, 0.91), s = 'underfitting', xytext=(18, 0.92), 
             arrowprops={'facecolor' : 'silver', 'width' : 4})

plt.show()
```
