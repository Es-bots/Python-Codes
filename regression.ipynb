# Regressione chand moteghayer

```python


from sklearn.datasets import load_boston
boston = load_boston()
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df['price'] = boston.target

boston_df
```

```python
x = boston.data 
y = boston.target 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size = 0.3, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
y_pred = reg.predict(x_test)
```

```python

plt.figure(figsize=(11, 11))
plt.scatter(y_test, y_pred)
plt.xlabel('Prices')
plt.ylabel('Predicted')
plt.show()
```

```python

from sklearn import metrics

mse = metrics.mean_squared_error(y_test, y_pred)
mse
```

```python
import cmath
cmath.sqrt(mse) # RMSE
```

```python

new_x = boston.data[:,[1,2]]
new_y = boston.target

new_x_train, new_x_test, new_y_train, new_y_test = train_test_split(
                    new_x, new_y, test_size = 0.3, random_state=42)

new_reg = LinearRegression()
new_reg.fit(new_x_train, new_y_train)
new_y_predict = new_reg.predict(new_x_test)

new_mse = metrics.mean_squared_error(new_y_test, new_y_predict)

new_mse
```

```python
cmath.sqrt(new_mse) # new RMSE
```

```python
coeff = np.array([1, 1, 2])

t = np.arange(-5,10)
y = np.polyval(coeff, t) + 20*np.sin(t)
plt.plot(t, y, 'bo-',color='red', markeredgecolor = 'blue')
plt.title("Train Data")
plt.xticks(t)
plt.yticks(y)


plt.grid()
```

```python


new_x_train = t
new_y_train = y

new_x_train = new_x_train.reshape(-1,1)
new_y_train = new_y_train.reshape(-1,1)

new_reg = LinearRegression()
new_reg.fit(new_x_train, new_y_train)
#new_y_predict = new_reg.predict(new_x_test)
plt.plot(new_x_train, new_reg.predict(new_x_train))
plt.show()
```

# Regression of not linear with RBF

```python
from sklearn.svm import SVR
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
y_rbf = svr_rbf.fit(new_x_train, new_y_train).predict(new_x_train)
plt.plot(new_x_train, y_rbf)
plt.show()
```

```python
metrics.mean_squared_error(new_y_train, y_rbf) 
```

# Regularization Regression

CostFunction = OLS + regularization term(Penalty), OLS: tabe hazine ghablieman

Penalizing large coefficients = Penalizing Overfitting
--------------------------------------------------------------------------

    jomleyi ke ezafe shode dar in ebarat karesh jarime kardane khat hayi ast ke meghdare 
    zarayebe bozorgi daranad, mikahyhim joloye entekhabe ye seri khat hara begirim.
    
    raveshe 1: ridge regression lost function = OLS(ordinary Least of squere) 
                                               + alfa * sigma(i = 1 ta n, a(i)**2)
    alfa: Hyperparameter, entekhabe aan tanzim mikone, kamo ziadesh bade.
    a(i): zarayebe Feature ha ke target ra misazad:
    y = a1X1 + a2X2 + ... + anXn + b, b = bios
    
    in kar baes mishe ke khat hayi entekhab beshan ke dar aan ha zaribe a koochektar ast,
    chon jarime kamtari be tabe hazineye ma ezafe mishe.
    
    in ravesh komak mikone zarayebe ma ta hade emkan koochik beshe, va Feature haye kam 
    ahamiat tar meghdaresh kamtar mishe, ta tasire kamtari dar modele ma dashte bashe,
    ama zaribi ra sefr nemikone, faghat tasire aan ra kam mikone.
    
    
    Large Coefficients => Overfitting, Overfitting baese pichidegie bishtare model mishe
    ba ghodrate tamime kamtar.
    
    raveshe 2: Lasso regression lost function = OLS(ordibary Least of squere) + alfa * 
                                                 sigma(i = 1 ta n, abs(a(i)))
    
    in ravesh baes mishe ke dar entekhabe zarayeb, zarayebe feature haye ezafi hazf 
    shavad, va dar vaghe feature hayi baghi mimanand ke tasire bishtari bar model, darand
    bara in ka pichidegie model kamtr mishavad, be in amal be estelah feature selection 
    gofte mishavad.
    
    masalan: target: ghodarate eghtesadi keshvar ha, feature: mizane saderat, rooz haye 
    aftabi
    
    mosalaman dar baresie ghodrate eghtesadie keshvar ha, tasire saderat kheyli bishtar 
    ast ta yek feature bi rabt mesle tedad rooz haye aftabi, Lasso Reg. in feature bi
    ahamiat ra tashkhis midahad, va hazf mikonad

```python


from sklearn.linear_model import Lasso
x = boston.data 
y = boston.target  
lasso = Lasso(alpha=0.1, normalize =True) 
lasso.fit(x, y)
lasso_coeff = lasso.coef_
print('x.shape = ', x.shape) 
```

```python
lasso_coeff 
```

```python

plt.figure(figsize=(11,11))
plt.plot(boston.feature_names, lasso_coeff)
plt.title('zarayebe Lasso, feature selection')
plt.show()
```
